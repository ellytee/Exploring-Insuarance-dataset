```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('http://www.ishelp.info/data/insurance.csv')
df.head()
```

```python
df.describe()
```

```python
print(f'sex: {df.sex.nunique()}')
print(f'region: {df.region.nunique()}')
print(f'smoker: {df.smoker.nunique()}')
print(f'age: {df.age.nunique()}')
```

```python
df.charges.std() #standard deviation
```

```python
#skew ness measuring  measuring if normality divates to the left or right found in scipy kurtosis measure the hight of normality or spread
from scipy.stats import skew, kurtosis

print(skew(df.charges, bias=False))
print(kurtosis(df.charges, bias=False))


```

```python
#to trust coeff  we need to follow the following  and caculate correlation
#-continuos data
#-linear relation ship make sure there is a linear trend 
#-homoskedastic raltionship area around regresion line is the same through out 
#p-value is probability that the results are nt goin to be repeated a low p value means will likely find the same value when you colect another dataset

df.charges.corr(df.bmi)
```

```python
corr_df= pd.DataFrame(columns=['r','p'])
for col in df:
    if pd.api.types.is_numeric_dtype(df[col]) and col != 'charges':
        r,p=stats.pearsonr(df.charges,df[col])
        corr_df.loc[col]=[round(r,3),round(p,3)]
corr_df
```

```python
plt.scatter(df.age,df.charges)
plt.title('insurance v charges')
plt.show()
```

```python
df_smoker=df[df['smoker']=='yes']
df_nonsmoker=df[df['smoker']=='no']
plt.scatter(df_smoker.age,df_smoker.charges, label='Smokers')
plt.scatter(df_nonsmoker.age,df_nonsmoker.charges, label='Non-Smokers')
plt.title('insurance v charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.legend()
plt.show()
```

```python
df_smoker_reduced=df_smoker.sample(50)
df_nonsmoker_reduced=df_nonsmoker.sample(50)
plt.scatter(df_smoker_reduced.age,df_smoker_reduced.charges, label='Smokers')
plt.scatter(df_nonsmoker_reduced.age,df_nonsmoker_reduced.charges, label='Non-Smokers')
plt.title('insurance v charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.legend()
plt.show()
```

```python
#making a line of regression
m, b, r, p, err=stats.linregress(df.age,df.charges)
x= range(18,65)
y= m*x +b
plt.plot(x,y)
plt.scatter(df.age,df.charges)
plt.title('insurance v charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
#heteroscedasticity show the spread of dots acros all values of x
#assumption is spread is equal in both ways
# the .fit

from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.diagnostic import het_white
from statsmodels.formula.api import ols
model = ols(formula='charges~age', data=df).fit()

white_test=het_white(model.resid, model.model.exog)
breuschpagen_test=het_breuschpagan(model.resid, model.model.exog)
output_df = pd.DataFrame(columns=['LM stat','LM p','F stat','F stat p'])
output_df.loc['White']= white_test
output_df.loc['Breusch-Pagen']= breuschpagen_test

output_df
```
